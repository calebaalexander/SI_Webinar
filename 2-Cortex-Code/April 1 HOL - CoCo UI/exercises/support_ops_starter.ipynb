# Support Ops Readiness Analysis
**PawCore SmartCollar V2 Launch Readiness**

This notebook analyzes whether the support team is operationally ready for V2 launch volume.
We will look at ticket patterns, customer sentiment, device telemetry, and build a composite readiness score.

```sql
-- Cell 1: Ticket Volume by Region
SELECT 
    region,
    COUNT(*) as total_tickets,
    COUNT(CASE WHEN LOWER(priority) = 'critical' THEN 1 END) as critical_tickets,
    ROUND(COUNT(CASE WHEN LOWER(priority) = 'critical' THEN 1 END) * 100.0 / COUNT(*), 1) as critical_pct
FROM PAWCORE_ANALYTICS.SUPPORT.SUPPORT_TICKETS
GROUP BY region
ORDER BY total_tickets DESC;
```

```sql
-- Cell 2: Customer Review Summary
-- BUG: Wrong schema name and wrong column name
SELECT 
    region,
    COUNT(*) as review_count,
    ROUND(AVG(stars), 2) as avg_rating,
    COUNT(CASE WHEN stars <= 2 THEN 1 END) as low_ratings
FROM PAWCORE_ANALYTICS.REVIEWS.CUSTOMER_REVIEWS
GROUP BY region
ORDER BY avg_rating ASC;
```

```sql
-- Cell 3: Device Telemetry - Low Battery Events
-- Returns no results due to overly restrictive filters
SELECT 
    region,
    COUNT(*) as low_battery_events,
    ROUND(AVG(battery_level), 2) as avg_battery_when_low
FROM PAWCORE_ANALYTICS.DEVICE_DATA.TELEMETRY
WHERE battery_level < 5
    AND temperature > 100
GROUP BY region
ORDER BY low_battery_events DESC;
```
